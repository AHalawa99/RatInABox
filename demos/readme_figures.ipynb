## Policy control 

```python
Env = Environment()
Ag = Agent(Env)

for i in tqdm(range(int(1.5 * 60 / Ag.dt))):
    drift = utils.rotate(Ag.pos - Env.centre, np.pi / 2)
    drift = 0.2 * drift / np.linalg.norm(drift)
    Ag.update(drift_velocity=drift, drift_to_random_strength_ratio=1)

fig, ax = Ag.plot_trajectory()
if save_plots == True:
    tpl.saveFigure(fig, "policycontrol")
```

## Polygon shaped environments, Environments with holes, curved Environments

```python
Env = Environment(params={"boundary": [[0, -0.2], [0, 0.2], [1.5, 0.5], [1.5, -0.5]]})
Ag = Agent(Env)

while Ag.t < 100:
    Ag.update()
fig, ax = Ag.plot_trajectory(t_end=45)
anim = Ag.animate_trajectory(t_start=10, speed_up=3)
if save_plots == True:
    # anim.save("../figures/animations/trapezium.mp4")
    tpl.saveFigure(fig, "trap_trajectory")
```

```python
Env = Environment(
    params={
        "aspect": 1.8,
        "holes": [
            [[0.2, 0.2], [0.8, 0.2], [0.8, 0.8], [0.2, 0.8]],
            [[1, 0.2], [1.6, 0.2], [1.6, 0.8], [1, 0.8]],
        ],
    }
)
Ag = Agent(Env)
while Ag.t < 150:
    if Ag.pos[0] < 0.9:  # anticlockwise on left
        drift = utils.rotate(Ag.pos - np.array([0.5, 0.5]), np.pi / 2)
    else:  # clockwise of right
        drift = utils.rotate(Ag.pos - np.array([1.3, 0.5]), -np.pi / 2)
    drift = 0.2 * drift / np.linalg.norm(drift)
    Ag.update(drift_velocity=drift, drift_to_random_strength_ratio=1)

anim = Ag.animate_trajectory(t_start=10, speed_up=5)
# fig, ax = Env.plot_environment()
fig1, ax1 = Ag.plot_trajectory(t_start=0, t_end=40)
if save_plots == True:
    # anim.save("../figures/animations/room_with_hole.mp4")
    # tpl.saveFigure(fig,"figure_of_eight")
    tpl.saveFigure(fig1, "figeight_trajectory")
```
