```python
from ratinabox.Neurons import FieldOfViewOVCells, ObjectVectorCells
from ratinabox import Agent
from ratinabox import Environment

#autoreload the libraries
%load_ext autoreload
%autoreload 2
```

```python
Env = Environment()
object_pos = Env.sample_positions(n=8,)
for i in range(8):
    Env.add_object(object_pos[i],type="same")
Env.add_wall([[0.5,0],[0.5,0.5]])

Ag = Agent(Env)
FoVs_OVC = FieldOfViewOVCells(Ag, params={
        "distance_range": [0.01, 0.4],  # min and max distances the agent can "see"
        "angle_range": [
            0,
            120,
        ],  # angluar FoV in degrees (will be symmetric on both sides, so give range in 0 (forwards) to 180 (backwards)
        "spatial_resolution": 0.04,  # resolution of each BVC tiling FoV
        "manifold_function": "hartley",
})


OVCs = ObjectVectorCells(Ag, 
                         params={
                            "distance_range": [0.01, 0.4],  # min and max distances the agent can "see"
                            "angle_range": [
                                0,
                                120,
                            ],  # angluar FoV in degrees (will be symmetric on both sides, so give range in 0 (forwards) to 180 (backwards)
                            "spatial_resolution": 0.04,  # resolution of each BVC tiling FoV
                            "manifold_function": "hartley",
                             })

for i in range(int(30/Ag.dt)):
    Ag.update()
    FoVs_OVC.update()
    OVCs.update()
    
```

```python
Env.plot_objects = True
Ag.animate_trajectory(t_end=30, speed_up=3, additional_plot_func=OVCs.display_manifold)
```
