```python
from ratinabox.Neurons import FieldOfViewOVCells
from ratinabox import Agent
from ratinabox import Environment

#autoreload the libraries
%load_ext autoreload
%autoreload 2
```

```python
Env = Environment()
object_pos = Env.sample_positions(n=8,)
for i in range(8):
    Env.add_object(object_pos[i],type="same")
Env.add_wall([[0.5,0],[0.5,0.5]])

Ag = Agent(Env)
FoVs_OVC = FieldOfViewOVCells(Ag)


for i in range(int(30/Ag.dt)):
    Ag.update()
    FoVs_OVC.update()
    
```

```python
Env.plot_objects = True
Ag.animate_trajectory(t_end=30, speed_up=3, additional_plot_func=FoVs_OVC.display_manifold)
```
