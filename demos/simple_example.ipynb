Initialise a 2D Environment. Initialise an Agent in the environment. Initialise some PlaceCells. Simulate for 20 seconds. Print table of times, position and firing rates and spikes. Plot trajectory and rate timeseries'.

```python
import ratinabox
from ratinabox.Environment import Environment
from ratinabox.Agent import Agent
from ratinabox.Neurons import *
```

```python
Env = Environment()
Ag = Agent(Env)
PCs = PlaceCells(Ag)
for i in range(int(20/Ag.dt)):
    Ag.update()
    PCs.update()
```

```python
print(Ag.history['t'][:10])
print(Ag.history['pos'][:10])
print(PCs.history['firingrate'][:10])
print(PCs.history['spikes'][:10])
```

```python
Ag.plot_trajectory()
PCs.plot_rate_timeseries()
```
