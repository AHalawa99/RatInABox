```python
import sys
sys.path.append("../")
import matplotlib.pyplot as plt 

%load_ext autoreload
%autoreload 2

import ratinabox
from ratinabox.Environment import Environment
from ratinabox.Agent import Agent
from ratinabox.Neurons import GridCells, PlaceCells
```

```python
Env = Environment(params={'dimensionality':'2D'})
Ag = Agent(Env)
Ag.speed_mean=0.3
# Ag.speed_mean = 0
Cs = PlaceCells(Ag)
while Ag.t < 300:
    Ag.update()
    Cs.update()
```

```python
fig, ax = plt.subplots(1,4,figsize=(12,6))
Ag.plot_trajectory(fig=fig,ax=ax[0],t_end=30)
Cs.plot_rate_map(method='groundtruth',fig=fig,ax=ax[1],chosen_neurons=1)
Cs.plot_rate_map(method='history',fig=fig,ax=ax[2],chosen_neurons=1)
Cs.plot_rate_timeseries(fig=fig,ax=ax[3],t_end=30)
```
